## training model

```python
all_done_fnb_tickets = authed_jira.search_issues('project = WSE AND status in (Done, Completed) AND (component in ("Support - FnB") OR "Customer Request Type" = "Support - FnB") AND component not in ("Ad hoc") ORDER BY updated DESC, created ASC, component ASC, priority DESC', maxResults=2000)
```

```python
len(all_done_fnb_tickets)
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
```

```python
sanitized_comments_array = []
sanitized_summary_array = []
```

```python
done_ticket_corpus = []
for done_ticket in all_done_fnb_tickets:
    ticket_dict = {}
    ticket_dict['jiraid'] = get_jira_id(done_ticket) 
    ticket_dict['title'] = get_title(done_ticket) 
    ticket_dict['summary'] = get_summary(done_ticket)
    list_of_comments = get_list_of_comments(done_ticket)
    ticket_dict['comments_data'] = [] 
    ticket_dict['comments_corpus'] = []
    for comment in list_of_comments:
        original_comment_data = {}
        original_comment_data['emailAddress'] = comment.author.emailAddress
        original_comment_data['body'] = comment.body
        original_comment_data['created'] = comment.created
        original_comment_data['updated'] = comment.updated
        ticket_dict['comments_data'].append(original_comment_data)
        comment_corpus_data = remove_code_from_comments(original_comment_data['body'])
        ticket_dict['comments_corpus'].append(comment_corpus_data)
    done_ticket_corpus.append(ticket_dict)
#     print('done for ',ticket_dict['jiraid'])
```

```python
document_collection_model_comments = [done_ticket['comments_corpus'] for done_ticket in done_ticket_corpus]
```

```python
document_collection_model_title = [(str(done_ticket['title'])+"\r\n "+str(done_ticket['summary'])) for done_ticket in done_ticket_corpus]
```

```python
tagged_data_mod1 = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(document_collection_model_title)]
```

```python
import nltk
nltk.download('punkt')
```

```python
max_epochs = 200
vec_size = 200
alpha = 0.025

model = Doc2Vec(size=vec_size,
                alpha=alpha, 
                min_alpha=0.025,
                min_count=1,
                dm =1)
  
model.build_vocab(tagged_data_mod1)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data_mod1,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v_title_mod1_200epoch_200vecsize.model")
print("Model Saved")
```

# Testing model

```python
test_issue = authed_jira.issue('WSE-2852')
test_issue_text_data = ''
test_issue_text_data += get_title(test_issue)
test_issue_text_data += get_summary(test_issue)
list_of_comments_final = get_list_of_comments(test_issue)
for comment in list_of_comments_final:
    test_issue_text_data += remove_code_from_comments(comment.body)
test_data = word_tokenize(test_issue_text_data.lower())
v1 = model.infer_vector(test_data)
print("V1_infer", v1)
# to find most similar doc using tags
similar_doc = model.docvecs.most_similar('5')
print(similar_doc)
```
